```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f
```

```python
# Define the degrees of freedom for the numerator (dfn) and denominator (dfd)
dfn = 5  # Degrees of freedom for the numerator
dfd = 10  # Degrees of freedom for the denominator

# Create an F-distribution object
f_dist = f(dfn, dfd)

# Generate some random data points from the F-distribution
sample_size = 1000
random_data = f_dist.rvs(size=sample_size)
```

```python
# Plot the histogram of the generated data points
plt.hist(random_data, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')

# Plot the probability density function (PDF) of the F-distribution
x = np.linspace(0, 5, 1000)
pdf = f_dist.pdf(x)
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

# Add labels and title to the plot
plt.xlabel('Random Variable')
plt.ylabel('Probability Density')
plt.title('F-Distribution (dfn = {}, dfd = {})'.format(dfn, dfd))
plt.legend()

# Show the plot
plt.show()
```
