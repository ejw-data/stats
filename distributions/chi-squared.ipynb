```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
# Define the degrees of freedom for the chi-squared distribution
degrees_of_freedom = 5

# Create a chi-squared distribution object
chi2_dist = chi2(df=degrees_of_freedom)

# Generate some random data points from the chi-squared distribution
sample_size = 1000
random_data = chi2_dist.rvs(size=sample_size)
```

```python
# Plot the histogram of the generated data points
plt.hist(random_data, bins=30, density=True, alpha=0.6, color='magenta', edgecolor='black')

# Plot the probability density function (PDF) of the chi-squared distribution
x = np.linspace(0, 20, 1000)
pdf = chi2_dist.pdf(x)
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

# Add labels and title to the plot
plt.xlabel('Random Variable')
plt.ylabel('Probability Density')
plt.title('Chi-Squared Distribution (Degrees of Freedom = {})'.format(degrees_of_freedom))
plt.legend()

# Show the plot
plt.show()
```
