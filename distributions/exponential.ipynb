```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

```python
# Define the scale (lambda) for the exponential distribution
scale = 1.5

# Create an exponential distribution object
exp_dist = expon(scale=scale)

# Generate some random data points from the exponential distribution
sample_size = 1000
random_data = exp_dist.rvs(size=sample_size)
```

```python
# Plot the histogram of the generated data points
plt.hist(random_data, bins=30, density=True, alpha=0.6, color='orange', edgecolor='black')

# Plot the probability density function (PDF) of the exponential distribution
x = np.linspace(0, 10, 1000)
pdf = exp_dist.pdf(x)
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

# Add labels and title to the plot
plt.xlabel('Random Variable')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution')
plt.legend()

# Show the plot
plt.show()
```
