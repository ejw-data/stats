```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
```

```python
# Define the degrees of freedom for the Student's t-distribution
degrees_of_freedom = 10

# Create a Student's t-distribution object
t_dist = t(df=degrees_of_freedom)

# Generate some random data points from the Student's t-distribution
sample_size = 1000
random_data = t_dist.rvs(size=sample_size)
```

```python
# Plot the histogram of the generated data points
plt.hist(random_data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the probability density function (PDF) of the Student's t-distribution
x = np.linspace(-5, 5, 1000)
pdf = t_dist.pdf(x)
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

# Add labels and title to the plot
plt.xlabel('Random Variable')
plt.ylabel('Probability Density')
plt.title("Student's t-distribution (Degrees of Freedom = {})".format(degrees_of_freedom))
plt.legend()

# Show the plot
plt.show()
```
